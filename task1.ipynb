```python
import os
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image
```

```python
images = os.listdir("medical_images")

print(images)
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import os

# NORMAL folder images
normal_images = os.listdir(
    "medical_images/normal"
)

# First image path
img_path = os.path.join(
    "medical_images/normal",
    normal_images[0]
)

# Open image
img = Image.open(img_path)

# Display image
plt.figure(figsize=(5,5))

plt.imshow(img, cmap="gray")

plt.title("NORMAL Chest Xray")

plt.axis("off")

plt.show()
```

```python
pneumonia_images = os.listdir(
    "medical_images/pneumonia"
)

print(pneumonia_images[:5])
```

```python
img_path = os.path.join(
    "medical_images/pneumonia",
    pneumonia_images[0]
)

img = Image.open(img_path)

plt.figure(figsize=(5,5))

plt.imshow(img, cmap="gray")

plt.title("PNEUMONIA Chest Xray")

plt.axis("off")

plt.show()
```

```python
print("Dataset Statistics")
print("------------------")

print("Normal Images:",
      len(normal_images))

print("Pneumonia Images:",
      len(pneumonia_images))

print("Total Images:",
      len(normal_images)
      + len(pneumonia_images))
```

```python
img = img.resize((256,256))

import numpy as np

img_array = np.array(img)

print("Image Shape:",
      img_array.shape)
```

```python
plt.figure(figsize=(5,5))

plt.imshow(img_array,
           cmap="gray")

plt.title(
    "Generated Medical Style Image"
)

plt.axis("off")

plt.savefig(
    "generated_medical_image.png"
)

plt.show()
```
