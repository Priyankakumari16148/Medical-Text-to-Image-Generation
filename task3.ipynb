```python
from transformers import BertTokenizer
from transformers import BertModel

import torch
import gradio as gr
```

```python
tokenizer = BertTokenizer.from_pretrained(
    "bert-base-uncased"
)

model = BertModel.from_pretrained(
    "bert-base-uncased"
)

print("BERT Loaded")
```

```python
text = "A chest Xray showing pneumonia"

print(text)
```

```python
tokens = tokenizer.tokenize(text)

print(tokens)
```

```python
token_ids = tokenizer.convert_tokens_to_ids(
    tokens
)

print(token_ids)
```

```python
inputs = tokenizer(
    text,
    return_tensors="pt"
)

print(inputs)
```

```python
with torch.no_grad():

    outputs = model(**inputs)

embeddings = outputs.last_hidden_state

print("Embedding Shape:")

print(embeddings.shape)
```

```python
print(embeddings)
```

```python
with open("embeddings_output.txt", "w") as f:
    f.write(str(embeddings))
```
